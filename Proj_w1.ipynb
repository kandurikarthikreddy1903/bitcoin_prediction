```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
#download bitcoin prices using requests
base_url = "https://api.gemini.com/v2"
response = requests.get(base_url + "/candles/btcusd/1day")
btc_data = pd.DataFrame(response.json(), columns =['time','open','high','low','close','volume'])
btc_data
```

```python
btc_data.to_csv("09_12_2023_btc_price.csv")
```
